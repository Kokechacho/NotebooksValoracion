# Estrategias con opciones

## 1. Introducci√≥n

En los mercados financieros, los Futuros y las Opciones est√°n ganando relevancia como productos potenciales para los inversionistas. Obtener m√°s informaci√≥n sobre estos instrumentos financieros es crucial para comprender su funcionamiento y c√≥mo aprovecharlos.

En Futuros y Opciones Financieras, Jaime D√≠az Tinoco y Fausto Hern√°ndez Trillo definen la Opci√≥n como un contrato que otorga al comprador el derecho, pero no la obligaci√≥n, de comprar o vender un activo a un precio y fecha predeterminados. Hay dos tipos de opciones: **las de Compra** (Call), donde el comprador tiene **el derecho a comprar** el activo, y **las de Venta** (Put), donde el comprador tiene **el derecho a venderlo**.

## 2. Arbitraje

El arbitraje con opciones financieras es **una estrategia de inversi√≥n** que aprovecha las disparidades de precios entre diferentes activos financieros para generar ganancias. Consiste en **adquirir un activo subvaluado y vender otro sobrevalorado** simult√°neamente para beneficiarse de la diferencia de precios. En el caso de las opciones financieras, esto implica la compra y venta simult√°nea de opciones de compra o venta con diferentes precios de ejercicio o fechas de vencimiento.

## 3. Tipos de arbitraje y ejercicios

### 3.1. CALL CUBIERTO

La estrategia de call cubierto consiste en** comprar acciones de un activo subyacente y vender simult√°neamente una opci√≥n de compra sobre ese mismo activ**o. Se utiliza cuando se espera que el precio del activo subyacente aumente ligeramente o se mantenga estable. Si el precio del activo subyacente sube, el comprador de la opci√≥n de compra puede ejercerla, pero el vendedor ya tiene las acciones subyacentes para vender. Si el precio no cambia, el vendedor conserva la prima recibida por vender la opci√≥n

Si el precio del activo subyacente no cambia, el comprador de la opci√≥n de venta pierde la prima que pag√≥ por la opci√≥n.

### 3.3. STRADDLE 

Una estrategia de straddle implica **la compra simult√°nea de una opci√≥n de compra y una opci√≥n de venta** en el mismo activo subyacente, con el mismo precio de ejercicio y la misma fecha de vencimiento. Se utiliza cuando se espera una volatilidad significativa en el precio del activo subyacente, pero no se sabe en qu√© direcci√≥n se mover√° el precio. Si el precio del activo subyacente aumenta por encima del precio de ejercicio de la opci√≥n de compra, el comprador de la opci√≥n de compra puede ejercerla y obtener ganancias. Si el precio del activo subyacente cae por debajo del precio de ejercicio de la opci√≥n de venta, el comprador de la opci√≥n de venta puede ejercerla y obtener ganancias. Si el precio del activo subyacente no cambia mucho, el comprador de ambas opciones pierde la prima que pag√≥ por las opciones.

#### EJERCICIO 3

Siendo nosotros el inversor n√∫mero 3 y viendo que los inversores anteriores no obtienen ganancias muy grandes dicidimos usar la estrategia con Stranddle.

Queremos invertir en bolsa tambi√©n en la URJC Corp, cuyo precio es de 50‚Ç¨ la acci√≥n, aunque como hay tantos inversores interesados √∫ltimamente en URJC Corp intuimos que dentro de poco ocurrir√° algo que modifique sustancialmente el precio de la acci√≥n.

Compramos simult√°neamente entonces una opci√≥n de compra y otra de venta, ambas a un precio de ejercicio de 50‚Ç¨ y con una prima de 3‚Ç¨ por acci√≥n.

**Calcula que escenario es m√°s rentable en base a las ganancias que te proporcione**

1. Escenario 1: URJC Corp sube en bolsa y supera los 56‚Ç¨ antes del vencimiento de la opci√≥n, ya que la demanda de m√∫ltiples inversores hizo que URJC se revalorizara. (¬øEjerceremos nuestras opciones?)
2. Escenario 2: URJC se desploma y no alcanza los 44‚Ç¨ debido a que tantas expectativas sobre URJC Corp lo convirti√≥ en una burbuja. (¬øEjerceremos nuestras opciones?)
3. Escenario 3: URJC Corp se mantiene estable y sigue rondando los 50‚Ç¨. (¬øEjerceremos nuestras opciones?)

<div style="background-color: #A9DFBF; padding: 10px;">
    <details style="color: black;">
        <summary>üí° Pista</summary>

    >> Si sube mucho el precio quiz√°s me interese ejercer el derecho de la opci√≥n de compra para obtener ganancias, mientras que no deber√≠a ejercer la venta de mis acciones ya que el precio de mercado est√° mucho m√°s alto.

*--Ejecuta el c√≥digo con Shift + Enter--*

```python
# ------------- NO MODIFICAR ESTE C√ìDIGO -------------------------
import numpy as np
import matplotlib.pyplot as plt

K_call = 50  # Precio de ejercicio de la opci√≥n de compra
K_put = 50   # Precio de ejercicio de la opci√≥n de venta
premium_call = 3  # Prima de la opci√≥n de compra
premium_put = 3   # Prima de la opci√≥n de venta

# Funci√≥n para calcular el payoff de una opci√≥n de compra
def payoff_call(S, K, premium):
    return np.maximum(S - K, 0) - premium

# Funci√≥n para calcular el payoff de una opci√≥n de venta
def payoff_put(S, K, premium):
    return np.maximum(K - S, 0) - premium

# Funci√≥n para calcular el payoff total del straddle
def payoff_straddle(S, K_call, K_put, premium_call, premium_put):
    return payoff_call(S, K_call, premium_call) + payoff_put(S, K_put, premium_put)

# Precios del activo subyacente
S_range = np.linspace(40, 60, 100)

# Calcular el payoff del straddle
payoff = payoff_straddle(S_range, K_call, K_put, premium_call, premium_put)
plt.figure(figsize=(10, 6))
```

```python
# ------------- AQUI VA TU C√ìDIGO -----------------------------
# GRAFICA QUE PASAR√çA EN EL ESCENARIO 1 -> PRECIO EN 60 
# GRAFICA QUE PASAR√çA EN EL ESCENARIO 1 -> PRECIO EN 42
# GRAFICA QUE PASAR√çA EN EL ESCENARIO 1 -> PRECIO EN 50.5
# PARA GRAFICAR LINEAS VERTICALES USAMOS:
# plt.axvline(x ...)
# EJEMPLO PARA GRAFICAR UNA LINEA EN X = 45 -----> plt.axvline(45)


# ------------- NO MODIFICAR ESTE C√ìDIGO -------------------------
# Graficar el payoff del straddle

plt.plot(S_range, payoff, label='Payoff del Straddle', color='blue', linestyle='-')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(K_call + premium_call, color='red', linestyle='--', label='Precio de ejercicio de la opci√≥n de compra + prima')
plt.axvline(K_put - premium_put, color='green', linestyle='--', label='Precio de ejercicio de la opci√≥n de venta - prima')
plt.xlabel('Precio del activo subyacente')
plt.ylabel('Payoff')
plt.title('Payoff de una estrategia de straddle')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.4. STRAGDLE 

Una estrategia de strangle es similar a la estrategia de straddle, pero **implica la compra simult√°nea de una opci√≥n de compra y una opci√≥n de venta en el mismo activo subyacente, con diferentes precios de ejercicio** y la misma fecha de vencimiento. Se utiliza cuando se espera una volatilidad significativa en el precio del activo subyacente, pero se espera que el precio se mueva en una direcci√≥n u otra. Si el precio del activo subyacente aumenta por encima del precio de ejercicio de la opci√≥n de compra, el comprador de la opci√≥n de compra puede ejercerla y obtener ganancias. Si el precio del activo subyacente cae por debajo del precio de ejercicio de la opci√≥n de venta, el comprador de la opci√≥n de venta puede ejercerla y obtener ganancias. Si el precio del activo subyacente no cambia mucho, el comprador de ambas opciones incurrir√° en p√©rdidas debido al costo de las primas pagadas.

#### EJERCICIO 4

En conclusi√≥n, siendo el √∫ltimo inversor, vamos a plagiar al inversor n√∫mero 3 y hacer esa misma estrategia pero cambiando los precios de las opciones.

Queremos invertir en bolsa tambi√©n en la URJC Corp, cuyo precio es de 50‚Ç¨ la acci√≥n.

Compramos simult√°neamente entonces una opci√≥n de compra y otra de venta a distintos precios:

a) Opci√≥n de compra: precio de 55‚Ç¨ y prima de 3‚Ç¨
b) Opci√≥n de venta: precio de 45‚Ç¨ y prima de 3‚Ç¨

**Modifica los par√°metros para ver como cambia el gr√°fico**

```python
# ------------- NO MODIFICAR ESTE C√ìDIGO -------------------------
import numpy as np
import matplotlib.pyplot as plt

# -------------- Modifica estos par√°metros --------------
K_call = 55  # Precio de ejercicio de la opci√≥n de compra
K_put = 45   # Precio de ejercicio de la opci√≥n de venta
premium_call = 3  # Prima de la opci√≥n de compra
premium_put = 3   # Prima de la opci√≥n de venta

# Funci√≥n para calcular el payoff de una opci√≥n de compra
def payoff_call(S, K, premium):
    return np.maximum(S - K, 0) - premium

# Funci√≥n para calcular el payoff de una opci√≥n de venta
def payoff_put(S, K, premium):
    return np.maximum(K - S, 0) - premium

# Funci√≥n para calcular el payoff total del strangle
def payoff_strangle(S, K_call, K_put, premium_call, premium_put):
    return payoff_call(S, K_call, premium_call) + payoff_put(S, K_put, premium_put)

# Precios del activo subyacente
S_range = np.linspace(40, 60, 100)

# Calcular el payoff del strangle
payoff = payoff_strangle(S_range, K_call, K_put, premium_call, premium_put)

# Graficar el payoff del strangle
plt.figure(figsize=(10, 6))
plt.plot(S_range, payoff, label='Payoff del Strangle', color='blue', linestyle='-')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(K_call + premium_call, color='red', linestyle='--', label='Precio de ejercicio de la opci√≥n de compra + prima')
plt.axvline(K_put - premium_put, color='green', linestyle='--', label='Precio de ejercicio de la opci√≥n de venta - prima')
plt.xlabel('Precio del activo subyacente')
plt.ylabel('Payoff')
plt.title('Payoff de una estrategia de strangle')
plt.legend()
plt.grid(True)
plt.show()
```
